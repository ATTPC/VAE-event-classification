```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
x = np.linspace(0, 1, 10)
lines = [1*x + b for b in range(10)]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
inferno = matplotlib.cm.get_cmap('inferno')
viridis = matplotlib.cm.get_cmap('viridis')

colors = [
    viridis(0.1),
    viridis(0.3),
    viridis(0.5),
    viridis(0.7),
    inferno(0.5),
    inferno(0.6),
    inferno(0.7),
    inferno(0.9)
]


for line, c in zip(lines, colors):
    ax.plot(np.arange(10), line, c=c, lw=3)
```
